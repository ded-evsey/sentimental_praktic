```python
import requests
import bs4
import csv
req= requests.get('https://www.rottentomatoes.com/m/deadpool_2/reviews/?type=top_critics')
FILENAME='parsingText.csv' 
#подключение библиотек, а так же подключение к сайту роттен томатос, для получения рецензий по новому фильму и 
#объявление переменной со значением пути к файлу, в котором будет храниться результат
```

```python
print(req)
#получение ответа от сайта
```

```python
parser = bs4.BeautifulSoup(req.text, 'html5lib')
print(parser)
#результат парсинга страницы
```

```python

with open(FILENAME, "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(parser.findAll('div', attrs={'class':'the_review'}))
    #Запись в файл данных результат парсинга, после поиска по диву с атрибутом class="the_review"
```

```python
with open(FILENAME, "r", newline="") as file:
    reader = csv.reader(file)
    for cel in reader:
        print(cel)
        #вывод из файла, для понимания того, что записывается в файл
```
